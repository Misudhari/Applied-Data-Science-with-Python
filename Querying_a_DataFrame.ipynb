### Boolean Masking is a heart of fast and efficient querying in pandas as well as numpy and its anlogous to bit masking is also used in other areas of computational science.
### Boolean masking is an array which can be of 1D like Series or 2D like DataFrames
### any cell with TRUE value will be fetched as a result and any cell with FALSE value will not be fetched.

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Datasets/Graduate School Admission Data/US_graduate_schools_admission_parameters_dataset.csv', index_col= 0)

df.columns = [x.lower().strip() for x in df.columns]

df.head()
```

```python
admit_mask = df['chance of admit'] > 0.7
admit_mask
```

```python
df.where(admit_mask).head()
```

```python
df.where(admit_mask).dropna().head()
```

```python
df[df['chance of admit'] > 0.7].head()
```

```python
df['gre score'].head()
```

```python
df[['gre score', 'toefl score']].head()
```

```python
df[df['gre score'] > 320].head()
```

#Multiple boolean Mask

```python
(df['chance of admit'] > 0.7) and (df['chance of admit'] < 0.9)
```

```python
(df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)
```

#Order of operations

```python
df['chance of admit'] > 0.7 & df['chance of admit'] < 0.9
```

```python
df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)
```

```python
df['chance of admit'].gt(0.7).lt(0.9)
```
