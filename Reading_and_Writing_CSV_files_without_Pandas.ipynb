#Data Files and Summary Statistics

```python
import csv

%precision 2

with open('/content/sample_data/california_housing_train.csv') as csvfile:
  housing_data = list(csv.DictReader(csvfile))

housing_data[:3]
```

```python
len(housing_data)
```

```python
housing_data[0].keys()
```

```python
sum(float(d['housing_median_age']) for d in housing_data) / len(housing_data)
```

```python
a = set(d['total_rooms'] for d in housing_data)
a
```

```python
b = []

for c in a:
  sumh_data = 0
  atypecount = 0
  for d in housing_data:
    if d['total_rooms'] == c:
      sumh_data += float(d['total_rooms'])
      atypecount += 1

b.append((c, sumh_data/atypecount))

b.sort(key = lambda x:x[0])

b
```
